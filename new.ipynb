```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv',index_col=1)
df
```

```python
df.head()
```

```python
new_columns= {
    "RH_1":'Kitchen Humidity', 
    "T1":'Kitchen Temp',
    "T2":'Lroom Temp',
    "T3":'Laundry Temp',
    "T4":'Office Temp', 
    "T5":'Bathroom Temp', 
    "T6":'Outside Temp',
    "T7":'IronRoom Temp', 
    "T8":'TeenRoom Temp', 
    "T9":'ParentRoom Temp',
    "RH_2":'Lroom Humidity', 
    "To":'Chievers Temp',
    "RH_4":'Office Humidity',
    "RH_3":'Laundry Humidity',
    "RH_out":'Humidity Outside', 
    "RH_5":'Bathroom Humidity', 
    "RH_6":'Outside Humidity',
    "RH_7":'IronRoom Humidity', 
    "RH_8":'TeenRoom Humidity', 
    "RH_9":'ParentRoom Humidity'}

df= df.rename(columns= new_columns, inplace= False)

df
```

```python
from sklearn.model_selection import train_test_split
y= df['Outside Temp']
x= df['Lroom Temp']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.3, random_state=7)
```

```python
x_test.shape
```

```python
x_train= x_train.values.reshape(-1,1)
```

```python
y_train= y_train.values.reshape(-1,1)
```

```python
x_test= x_test.values.reshape(-1,1)
```

```python
y_test= y_test.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(x_train, y_train)
pred_value = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score= r2_score(y_test,pred_value)
round(r2_score,2)
```
